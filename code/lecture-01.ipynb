```python
# For plots
import matplotlib.pyplot as plt
import numpy as np
```

# 1. Linear Algebra

```python
# Demo for eigen decomposition

from sklearn.datasets import make_spd_matrix
from numpy.linalg import eig

P = make_spd_matrix(5)
Lambda, U = eig(P)

print("Eigen values:")
print(Lambda)

R = np.matmul(U.T,U)
print("\nVerify that U is an orthogonal matrix")
print(R)
```

# 2. Probability Distribution

# 2.1 Sampling from Gaussian Distributions

```python
# https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.random.multivariate_normal.html
from numpy.random import multivariate_normal as mv_normal
```

```python
# Define a two-dimensional Gaussian by specifying its mean and covariance
# Mean and Covariance

mean = [0,0]
cov = [[1,0],[0,1]]
N = 500 # Number of samples
x1, x2 = mv_normal(mean, cov, N).T
plt.plot(x1, x2, 'o')
plt.xlim(-4,4)
plt.ylim(-4,4)
```

```python
# Define another Gaussian with non-identical covariance matrix

mean = [0,0]
cov = [[2,0],[0,1]]
N = 500 # Number of samples
x1, x2 = mv_normal(mean, cov, N).T
plt.plot(x1, x2, 'o')
plt.xlim(-4,4)
plt.ylim(-4,4)
```
